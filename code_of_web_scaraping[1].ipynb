```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import sys
sys.setrecursionlimit(10000)
sys.getrecursionlimit()
```

```python
# #url_df=pd.read_csv('urls.csv')
url_df=pd.read_csv(r"productURLs.csv")
url_df
```

```python

def get_product_info(url,headers,df,df_n):
    try:
      html = requests.get(url, headers=HEADERS)
      soup = BeautifulSoup(html.text,"html.parser")
      id =  url.split("/")[-2]
      #title = soup.find('span', {'id':"productTitle"}).text.strip()
      title= url.split("/")[-4]
      price = soup.find('span', {'class':"a-offscreen"}).text.strip()
      rating = soup.find("span", attrs={'class': 'a-icon-alt'}).text.strip()
      images = re.findall('"hiRes":"(.+?)"', html.text)
      description = soup.find('div', {'id': "feature-bullets"}).text.strip()
      df=df.append(
          {
              'product_id':id,
              'product_title' : title , 
              'product_price ' : price, 
              'product_rating' :re.split('\s+', str(rating))[0],
              'product_images':images,
              'product_description':description
              }, ignore_index = True)
        #print("appended "+id )
        #print(title, "\n", price, "\n",rating,  "\n", description, "\n", images)
    except:
      df_n=df_n.append({'url':url}, ignore_index = True)
      #print(url)
    #print("done")
    return df,df_n
    
```

```python

HEADERS = {
    'User-Agent': ('Mozilla/5.0 (X11; Linux x86_64)'
                    'AppleWebKit/537.36 (KHTML, like Gecko)'
                    'Chrome/44.0.2403.157 Safari/537.36'),
    'Accept-Language': 'en-US, en;q=0.5'
}
product_details=pd.DataFrame()
err_product_url=pd.DataFrame()
for i in range(int(url_df.count())):
    product_details,err_product_url=get_product_info(str(url_df.iloc[i, 0]),HEADERS,product_details,err_product_url)
    print(str(i+1)+'/'+ str(re.split('\s+', str(url_df.count()))[1]))
product_details
```
